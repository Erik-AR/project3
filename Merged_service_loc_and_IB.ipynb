```python
import pandas as pd
```

```python
Install_Base = "IB_complete_data.csv"
Service_loc = "Service_locations.csv"
```

```python
Install_Base_df = pd.read_csv(Install_Base)
Service_loc_df = pd.read_csv(Service_loc)
```

```python
Install_Base_df.keys()
Install_Base_df.count()
```

```python
# Remove the rows with missing data
clean_Install_Base_df = Install_Base_df.dropna(how="any")
clean_Install_Base_df.count()
```

```python
Service_loc_df.head()
```

```python
Service_merge_df =Service_loc_df.rename(columns={"SERIAL_NO":"SERIAL_NUMBER"})
Service_merge_df.count()
```

```python
Clean_Service_merge_df = Service_merge_df.dropna(how="any")
Clean_Service_merge_df.count()
```

```python
merged_table = pd.merge(clean_Install_Base_df, Clean_Service_merge_df, on='SERIAL_NUMBER', how="left")
merged_table
```
